### Анализ самых опасных месяцев. / Analysis of the most dangerous months.

```python
query=''' SELECT DISTINCT to_char,count FROM (SELECT DISTINCT to_char(collision_date::date,'MONTH'),
        EXTRACT(YEAR FROM collision_date::date),
        COUNT(case_id) OVER(PARTITION BY EXTRACT(MONTH FROM collision_date)) 
FROM collisions) as m
        WHERE date_part in (SELECT date_part FROM  (SELECT date_part, COUNT(*) AS num_months
FROM (
    SELECT EXTRACT(YEAR FROM collision_date::date) AS date_part,
           EXTRACT(MONTH FROM collision_date::date) AS month
    FROM collisions
    GROUP BY date_part, month
) AS m
GROUP BY date_part
HAVING COUNT(*) = 12) as y)
        ORDER BY count DESC
'''
```

```python
crash_month=read(query)
plt.figure(figsize=(14,8))
plt.plot(crash_month['to_char'],crash_month['count'],'o-')

plt.grid()
plt.xlabel('Месяц аварии')
plt.ylabel('Количество аварий')
plt.title('График аварий по месяцам')
```

```python
query=''' SELECT s.VEHICLE_TYPE,c.COLLISION_DAMAGE as type_of_collision,count(s.CASE_ID)
          FROM Vehicles as s

          JOIN collisions as c ON c.case_id=s.case_id
          WHERE c.COLLISION_DAMAGE is not Null
          GROUP By s.VEHICLE_TYPE,c.COLLISION_DAMAGE
          
          ORDER BY count DESC
'''
data=read(query).pivot_table(index='vehicle_type',columns='type_of_collision',values='count',aggfunc='sum')
data.plot(grid=True,style='o-',figsize=(12,8))
plt.xticks(rotation='vertical');
plt.title('Какой тип кузова чаще получает урон')
```

## Создайте модель для оценки водительского риска / Create a model for driver risk assessment

### Создание таблицы необходимой для оубчения модели. / Creating a table necessary for training the model.

```python
query=''' SELECT c.*,p.*,v.* FROM collisions as c
        JOIN Parties as p ON p.case_id=c.case_id
        JOIN Vehicles as v ON v.case_id=c.case_id
        WHERE p.PARTY_TYPE='car' AND (c.COLLISION_DAMAGE!='scratch' AND EXTRACT(YEAR FROM c.collision_date)='2012')
'''
df=read(query)
```

### Создание модели и предобработка. / Model creation and preprocessing.

**Метрика :**
Учитывая условие задачи то я считаю *Recall* идеально сюда подойдет как основная метрика.Так как нам важно обязательно определить человека который может совершить дтп ,даже если человек может быть не виноват.

ENG:

Metric : Given the condition of the task, I think Recall is ideally suited here as the main metric.Since it is important for us to identify the person who can commit an accident, even if the person may not be to blame.

```python
features=df_new.drop('at_fault',axis=1)
target=df_new['at_fault']
```

```python
ohe=['intersection']
oe=[i for i in features.columns if i not in ohe and features[i].dtypes=='object']

preprocesing=make_column_transformer((OrdinalEncoder(handle_unknown='ignore'), oe),
                                    (OneHotEncoder(handle_unknown='ignore'), ohe),
                                    (MinMaxScaler(),[i for i in features.columns if features[i].dtypes!='object']))


print('Размер перед разделением :',features.shape)

features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.25,stratify=target,shuffle=True,random_state=1324)

print('Размер после  разделения обучающая выборка :',features_train.shape)
print('Размер после разделения тестовая выборка :',features_test.shape)


feat=[features_train,features_test]
tar=[target_train,target_test]
```

#### Функция обучения / Study function

```python
import time

def study(features, target, model, params):
    
    pline = Pipeline([('preprocessing', preprocesing),
                      ('model', model)])
    
    grid = RandomizedSearchCV(pline, params, cv=2, scoring='recall')
    
    start = time.time()  # Start timer
    grid.fit(features[0], target[0])
    end = time.time()  # End timer
    
    time_taken = end - start
    
    print(f'Time taken for hyperparameter search: {time_taken} seconds')
    
    print(f'Recall-train: {grid.best_score_}')
    
    return grid.best_estimator_, grid.cv_results_, time_taken
```

#### Модели / Models

#### Лес / Forest

```python
params={'model__n_estimators': range(20,200,10), 'model__max_depth': range(5,50,2),
        'model__min_samples_split': [5,2], 'model__min_samples_leaf': [1,5], 'model__bootstrap': [True,False]}
forest,forest_result,time_forest=study(feat,tar,RandomForestClassifier(random_state=1224),params)
```

#### Ridge

```python
param_grid_ridge = {
    'model__alpha': [0.1, 0.01],  
    'model__fit_intercept': [True, False], 
    'model__normalize': [True, False],  
    'model__class_weight': [None, 'balanced']  
}

ridge,results_ridge,ridge_time=study(feat,tar,RidgeClassifier(random_state=12234),param_grid_ridge)
```

#### LightGBM

```python

param_grid = {
    
    'model__learning_rate': np.arange(0.01,0.1,0.01),  
    'model__n_estimators': range(100, 1000), 
    'model__max_depth': range( 1, 10, 1 ),  
    'model__min_split_gain': [ 0.1, 0.2,],  
    'model__subsample': [0.5, 0.6]  

}
light,light_result,light_time=study(feat,tar,LGBMClassifier(random_state=1224),param_grid)
```

#### Catboost

```python
param_grid_cb = {
    'model__iterations': range(100, 1000,100),
    'model__learning_rate': np.arange(0.01,0.1,0.01),
    'model__depth': range( 1, 10, 1 ),
    'model__l2_leaf_reg': [1, 3]}
cat,cat_result,cat_time=study(feat,tar,CatBoostClassifier(random_state=1224,verbose=0),param_grid_cb)
```
